```python
import os
import sys
import numpy as np
sys.path.append(os.path.abspath('.')+'/../ampyl')
import ampyl
```

```python
def get_value_direct(E, nP, L, kellm_space_row, kellm_space_col):
    G_direct = [[]]
    for kellm_entry_row in kellm_space_row:
        G_row = []
        for kellm_entry_col in kellm_space_col:
            kentry_row = kellm_entry_row[:3]
            kentry_col = kellm_entry_col[:3]
            ell_row = kellm_entry_row[3]
            mazi_row = kellm_entry_row[4]
            ell_col = kellm_entry_col[3]
            mazi_col = kellm_entry_col[4]
            Gtmp = ampyl.QCFunctions.getG_single_entry(E=E,
                nP=nP,
                L=L,
                np1spec=kentry_row,
                np2spec=kentry_col,
                ell1=ell_row,
                mazi1=mazi_row,
                ell2=ell_col,
                mazi2=mazi_col,
                m1=1.0,
                m2=1.0,
                m3=1.0,
                alpha=-1.0,
                beta=0.0,
                J_slow=False,
                three_scheme='relativistic pole',
                qc_organization='hermitian, real harmonics')
            G_row = G_row+[Gtmp*fcs.g_templates[0][0][0][0]]
        G_direct = G_direct+[G_row]
    G_direct = np.array(G_direct[1:])
    return G_direct
```

## Explicit, P=[000]

```python
fc = ampyl.FlavorChannel(3)
print(fc)
fcs = ampyl.FlavorChannelSpace(fc_list=[fc])
fvs = ampyl.FiniteVolumeSetup()
tbis = ampyl.ThreeBodyInteractionScheme()
qcis = ampyl.QCIndexSpace(fcs=fcs, fvs=fvs, tbis=tbis, Emax=5.0, Lmax=7.0, verbosity=2)
```

```python
g = ampyl.G(qcis=qcis)
```

```python
G = g.get_value(5.0, 7.0)
```

```python
G.shape
```

```python
len(qcis.kellm_spaces[0])
```

```python
qcis.kellm_spaces[0][0].shape
```

```python
G_direct = get_value_direct(E=5.0, nP=np.array([0, 0, 0]), L=7.0,
                            kellm_space_row=qcis.kellm_spaces[0][0],
                            kellm_space_col=qcis.kellm_spaces[0][0])
```

```python
(G-G_direct < 1.0e-15).all()
```
